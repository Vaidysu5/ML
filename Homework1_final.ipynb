```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
mlmdata=pd.read_csv('ml1mdata2000.csv')
mlmdata['date']=pd.to_datetime(mlmdata['timestamp'],unit='s',errors='coerce')
mlmdata['year']=pd.DatetimeIndex(mlmdata['date']).year
mlmdata['month']=pd.DatetimeIndex(mlmdata['date']).month
```

```python
unique_years = mlmdata["date"].dt.year.unique()
print(unique_years)
#Q1. Unique years featured in dataset
```

```python
mlmdata.head(10)
#printing first 10 rows of dataset
```

```python
value_counts = mlmdata['year'].value_counts()
print(value_counts)
#Q2. number of records for each year
```

```python
year_counts = mlmdata.groupby(pd.to_datetime(mlmdata['date']).dt.year)['title'].nunique()
print(year_counts)
#Q2. number of unique records for each year (after grouping the data based on unique titles)
```

```python
ratings = mlmdata["rating"].iloc[:10]
plt.plot('ratings')
plt.xlabel("Observation Number (1-10)")
plt.ylabel("Movie Rating")
plt.title("Movie Ratings from Observation 1 to 10")
plt.grid(True)
plt.show()

#Q5. No clear trend: The plot doesn't exhibit a distinct upward or downward trend, 
#suggesting the ratings remain relatively stable within this range.
#The ratings consistently hover around 4 or 5, suggesting overall positive reception
```

```python
mlmdata_filtered = mlmdata.head(100)
plt.figure(figsize=(8, 6)) 
plt.scatter(mlmdata_filtered["rating"], mlmdata_filtered["age"], alpha=0.7)
plt.xlabel("Movie Rating")
plt.ylabel("Age")
plt.title("Movie Ratings vs. Age (First 100 Observations)")
plt.show()

#Q6. The ratings consistently hover around 4 or 5 for all age groups, 
# suggesting overall positive reception for the movie
# The ratings are clustered around 4 or 5 for all age groups as represented by the chart
```
