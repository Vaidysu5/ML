```python
import os
import numpy as np
import pandas as pd
import math
import matplotlib.pylab as plt
from IPython.display import display
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn import ensemble
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

df_mobile = pd.read_csv ("mobile.csv")
Mobile=df_mobile.dropna()
Mobile.head()
```

```python
Mobile.dtypes
```

```python
X=Mobile.drop(['price_range'],axis=1)
Y=Mobile['price_range']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.70, random_state=1)
print("X_train shape: {}". format(X_train.shape))
print("Y_train shape: {}". format(Y_train.shape))
```

```python
print("X_test shape: {}". format(X_test.shape))
print("Y_test shape: {}". format(Y_test.shape))
```

```python
clf = DecisionTreeClassifier(criterion="gini", max_depth=3)
clf.fit(X_train, Y_train)

plt.figure(figsize=(20,8))  # set plot size (denoted in inches)
tree.plot_tree(clf, feature_names=X.columns, class_names=['0','1','2','3'], filled=True, fontsize=12)

#Q1a(i)decision tree classifier that uses gini index as the impurity measure with a maximum depth of 3
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_test, clf.predict(X_test))
print("Accuracy on the test set:", accuracy)

#Q1a(ii) accuracy on test set
```

```python
def evaluate_decision_tree(max_depth):
    # Train the decision tree classifier
    clf = DecisionTreeClassifier(criterion="gini", max_depth=max_depth)
    clf.fit(X_train, Y_train)
    train_accuracy = clf.score(X_train, Y_train)
    test_accuracy = clf.score(X_test, Y_test)
    return train_accuracy, test_accuracy
```

```python
depths = range(2, 20)
train_accuracies, test_accuracies = [], []
for depth in depths:
    train_accuracy, test_accuracy = evaluate_decision_tree(depth)
    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)
    
##Q1a(iii) Accuracies for different depth values
```

```python
   
plt.figure(figsize=(10, 6))
plt.plot(depths, train_accuracies, label="Training Accuracy")
plt.plot(depths, test_accuracies, label="Test Accuracy")
plt.xlabel("Max Depth")
plt.ylabel("Accuracy")
plt.title("Decision Tree Accuracy vs. Depth")
plt.legend()
plt.grid(True)
plt.show()

## Q1a(iii) As the depth of the decision tree increases, the training accuracy increases, while the test accuracy 
## is flatter and starts to plateau as overfitting starts to take charge 
## This suggests that the model is overfitting the training data and not generalizing well to unseen data.
```

```python
def evaluate_knn(k):
    # Train the k-NN classifier
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(X_train, Y_train)

    # Predict on training and test sets
    Y_train_pred = clf.predict(X_train)
    Y_test_pred = clf.predict(X_test)

    # Calculate accuracy scores
    accuracy_train = accuracy_score(Y_train, Y_train_pred)
    accuracy_test = accuracy_score(Y_test, Y_test_pred)

    return accuracy_train, accuracy_test
```

```python
k_values = range(2, 15)
train_accuracies, test_accuracies = [], []
for k in k_values:
    accuracy_train, accuracy_test = evaluate_knn(k)
    train_accuracies.append(accuracy_train)
    test_accuracies.append(accuracy_test)
    
#Q1(b)k-nn classifier for different number of nearest neighbors and their accuracies
```

```python
plt.figure(figsize=(10, 6))
plt.plot(k_values, train_accuracies, label="Training Accuracy")
plt.plot(k_values, test_accuracies, label="Test Accuracy")
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.title("k-NN Accuracy vs. Number of Neighbors")
plt.legend()
plt.grid(True)
plt.show()

#Q1(b)k-nn classifier for different number of nearest neighbors and their accuracies
```

```python
print("Analyzing results:")
# Find the k with the highest test accuracy
best_k = k_values[test_accuracies.index(max(test_accuracies))]
print(f"Best k (highest test accuracy): {best_k}")
print(f"Test accuracy with k={best_k}: {max(test_accuracies)}")

#Q1(b)k-nn classifier for different number of nearest neighbors and their accuracies
```

```python
C_values = [0.001, 0.01, 0.1, 1]
train_accuracies = []
test_accuracies = []


for C in C_values:
    model = SVC(kernel='linear', C=C)
    model.fit(X_train, Y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    train_accuracy = model.score(X_train, Y_train)
    test_accuracy = model.score(X_test, Y_test)
    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)
    
plt.plot(C_values, train_accuracies, label='Training Accuracy')
plt.plot(C_values, test_accuracies, label='Test Accuracy')
plt.xlabel('C Value')
plt.ylabel('Accuracy')
plt.title('Linear SVM Accuracy with Different C Values')
plt.legend()
plt.show()

#Q1(c) support vector machine with a linear kernel and accuracies for different C values
```

```python
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier

n_estimators = 150
max_depth = 5
models = {
    'Random Forest': RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth),
    'Bagging': BaggingClassifier(base_estimator=DecisionTreeClassifier(max_depth=max_depth), n_estimators=n_estimators),
    'AdaBoost': AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=max_depth), n_estimators=n_estimators)
}

for model_name, model in models.items():
    model.fit(X_train, Y_train)


train_accuracies = {}
test_accuracies = {}
for model_name, model in models.items():
    train_accuracies[model_name] = model.score(X_train, Y_train)
    test_accuracies[model_name] = model.score(X_test, Y_test)

index = range(len(models))
bar_width = 0.35
x_train_pos = [i - bar_width / 2 for i in index]
x_test_pos = [i + bar_width / 2 for i in index]

plt.figure(figsize=(8, 6))
plt.bar(x_train_pos, train_accuracies.values(), bar_width, label='Training Accuracy')
plt.bar(x_test_pos, test_accuracies.values(), bar_width, label='Test Accuracy')
plt.xticks([i + bar_width / 2 for i in index], train_accuracies.keys())
plt.xlabel('Ensemble Method')
plt.ylabel('Accuracy')
plt.title('Accuracy of Ensemble Classifiers')
plt.legend()
plt.tight_layout()
plt.show()

best_model_name = max(train_accuracies, key=train_accuracies.get)
print(f"Best performing model (training): {best_model_name}")
best_model_name = max(test_accuracies, key=test_accuracies.get)
print(f"Best performing model (test): {best_model_name}")

#Q1(d): ensemble classifiers with best performing model
```

```python
df_housing = pd.read_csv ("data1.csv")
df_housing.head()
```

```python
Housing = df_housing.drop(columns=["housing_median_age", "total_rooms", "total_bedrooms","population", "households", "median_house_value", "ocean_proximity"], axis=1)

Housing.head()

```

```python
from sklearn import cluster


k_means = cluster.KMeans(n_clusters=8, max_iter=50, random_state=1)
k_means.fit(Housing) 
labels = k_means.labels_
pd.DataFrame(labels, index=Housing.index, columns=['Cluster ID'])

#Q2(a) kclusters to its nearest centroid
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

numClusters = [1,2,3,4,5,6,7,8,15]
SSE = []
for k in numClusters:
    k_means = cluster.KMeans(n_clusters=k)
    k_means.fit(Housing)
    SSE.append(k_means.inertia_)

plt.plot(numClusters, SSE, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')

#Q2(c) the plot looks like an arm with a clear elbow at k = 2.
```

```python
X = StandardScaler().fit_transform(Housing)
def calculate_silhouette_scores(X, k_range):
  silhouette_scores = []
  for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(X, labels))
  return silhouette_scores

k_range = range(2, 11)
print("Silhouette Scores for different K values:")
for k, score in zip(k_range, silhouette_scores):
  print(f"K = {k}: {score:.4f}")



#Q2(c) higher silhouette scores indicate better cluster separation and K=2 has the highest silhoutte score
```
