```python
# Model to predict winner between two teams
```

```python
df_IPL = df_IPL.dropna(subset=['winner'])
```

```python
features = ["team1", "team2","toss_winner","toss_decision","venue"]
target = ["winner"]
```

```python
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore') 
encoded_features = pd.DataFrame(encoder.fit_transform(df_IPL[features]))
```

```python
X_train, X_test, y_train, y_test = train_test_split(encoded_features, df_IPL['winner'], test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
new_match = pd.DataFrame({'team1': ['Mumbai Indians'], 
                          'team2': ['Chennai Super Kings'], 
                          'toss_winner': ['Mumbai Indians'], 
                          'toss_decision': ['field'],
                         'venue':['MA Chidambaram Stadium, Chepauk']})
encoded_new_match = encoder.transform(new_match) 

predicted_winner_code = model.predict(encoded_new_match)[0]

print(f"Predicted Winner (Encoded): {predicted_winner_code}")
```

```python
city_team_map = {
    "Hyderabad": "Sunrisers Hyderabad",
    "Pune": "Rising Pune Supergiant",
    "Rajkot": "Gujarat Lions",
    "Bangalore": "Royal Challengers Bangalore",
    "Mumbai": "Mumbai Indians",
    "Kolkata": "Kolkata Knight Riders",
    "Delhi": "Delhi Daredevils",
    "Chandigarh": "Kings XI Punjab",
    "Jaipur": "Rajasthan Royals",
    "Chennai": "Chennai Super Kings",
    "Ahmedabad": "Gujarat Lions",
    "Kochi": "Kochi Tuskers Kerala",
    "Visakhapatnam": "Sunrisers Hyderabad"
}

df_IPL = df_IPL.dropna(subset=['winner'])

def get_home_ground(row):
    city, winner = row['city'], row['winner']
    return city_team_map.get(city, None) == winner and "Home Win" or "Away Win"

df_IPL['Home Ground'] = df_IPL.apply(get_home_ground, axis=1)
df_IPL.head(5)
```

```python
team_wins = df_IPL.groupby("winner")["Home Ground"].value_counts().unstack(fill_value=0)


colors = ["skyblue", "lightcoral"]


plt.figure(figsize=(30, 20))  # Adjust width and height as needed
team_wins.plot(kind="bar", stacked=True, color=colors)


plt.xlabel("Team", fontsize=12)
plt.ylabel("Number of Wins", fontsize=10)
plt.title("Home vs. Away Win Split for Each Team", fontsize=14)


plt.legend(title="Location")


plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
team_wins = df_IPL.groupby("winner")["Home Ground"].value_counts().unstack(fill_value=0)


team_wins["Total"] = team_wins.sum(axis=1)


for col in team_wins.columns:
  if col != 'Total':
    team_wins[col + "%"] = (team_wins[col] / team_wins["Total"]) * 100

print(team_wins.to_string())
```

```python


umpire_wins = df_IPL.groupby("umpire1")["winner"].value_counts().unstack(fill_value=0)


max_wins_per_umpire = umpire_wins.max(axis=1)

print(f"Maximum Wins for Each Umpire:")
for umpire, wins in max_wins_per_umpire.items():
  print(f"- Umpire: {umpire}, Maximum Wins: {wins}")
```

```python
umpire_wins = df_IPL.groupby('umpire1')['winner'].value_counts().unstack()


top_5_umpires = umpire_wins.sum(axis=1).sort_values(ascending=False).head(5).index.to_numpy()

top_5_teams = umpire_wins.loc[top_5_umpires]

print("Top 5 Umpires and Winning Teams:")
print(pd.concat([pd.Series(top_5_umpires, name='Umpire'), top_5_teams], axis=1))
```
